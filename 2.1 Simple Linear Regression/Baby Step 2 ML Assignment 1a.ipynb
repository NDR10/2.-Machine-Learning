```python
# Pandas is a library that contains codes to handle tables in a Python file. 
# In the below code the entire pandas library is called and a new name "pd" is assigned to the pandas library

import pandas as pd
```

```python
# A variable named "dataset" is created and the table "Salary_Data.csv" is linked to that variable. 
# It should be ensured that the table "Salary_Data.csv" and the Python file are in the same folder.
# It should be noted the table has an extension .csv, if it has some other extension. We can use that in the code.
# Then the created variable is called using the second line of the code

dataset = pd.read_csv("Salary_Data.csv")
dataset
```

```python
# In this specific cls file, there are two columns. The first column is considered as the input. The second column is output.
# The first column's values in the dataset are linked to a variable "Input" using the code below.
# If there are more than one inputs then the code can be written as Input = dataset[["Column1name", "Column2name", "Column3name", "etc"]]
# It should be noted that assigning a part of dataset to the variable input is possible by using two square brackets for opening and also closing
# The input is then called 

Input = dataset[["YearsExperience"]]
Input
```

```python
# The above procedure is repeated for the output
# The output is then called

Output = dataset[["Salary"]]
Output
```

```python
# Similar to pandas there is another library named as "sklearn". This library contains a set of codes that can help in producing ML in Python
#The sklearn contains many classes. The code "sklearn.model_selection" is used to call the sub-class "model_selection" from the "sklearn" library
# "train_test_split" is a function that is available in the "model_selection" from the "sklearn" library
# "train_test_split"  helps to split a set of data from the "Salary_Data.csv" for training and testing respectively

from sklearn.model_selection import train_test_split
```

```python
# The LHS of the given code is used to assign 70% of data from "Salary_Data.csv" as the training data set and the remaining as a test data set
# It should be noted that the data set for input is considered as X and the output value is assigned to Y
# The RHS of the code is a parametrized function
# THis code assigns values of input and output for the training data set and testing data set
# LIkewide the size of the test data set is set as 30 %
# The random value is set as 0

X_train, X_test, Y_train, Y_test = train_test_split(Input, Output, test_size = 0.3, random_state = 0)
```

```python
# "linear_model" is a subclass inside the "sklearn" library
# "LinearRegression" is a subclass that is inside "linear_model" that contains codes to calculate regression for a linear dataset
# The variable "Regressor" is linked to the non-parametrised class i.e., "LinearRegression()"

from sklearn.linear_model import LinearRegression
Regressor = LinearRegression()
Regressor.fit(X_train, Y_train)
```

```python
# The weight of the linear regression equation is calculated using the function coef_
# The weight is then called using the second line
# It is noted that the value of the weight is "9360"

Weight = Regressor.coef_
Weight
```

```python
# The bias of the linear regression equation is calculated using the function intercept_
# The bias is then called using the second line
# It is noted that the value of the bias is "2677"

Bias = Regressor.intercept_
Bias
```

```python
# There is a function named "predict" in the "LinearRegression" class
# The variable "Y_pred" is linked to the input value of the test dataset

Y_pred = Regressor.predict(X_test)
```

```python
# There is a function called "r2_score" in the subclass "metrics" in the "sklearn" library
# The variable "r_score" is linked to the parametrized function "r2_score". 
# In this case the "Y_test and Y_pred" are the parametics for the "r2_score"
# The variable "r_score" is then called
# # It is noted that the value of the regression is "0.9741"
# Since the regression value is closer to 1, we can consider that the model is having good fit. Hence, this model will perform well during the test.

from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
r_score
```

```python
# The pickle is a library that helps to save ML programs written using Python codes
# A variable "filename" is created. This is given the name of the file with an extension .av to save the file using the "pickle" library
# There is a class "dump" in the "pickle" library. THis helps to save the ML program.
# The third line of the code creates the file "Python_program_Linear_Model.sav" in the same folder that contains "Salary_Data.csv"
# Note: "wb" means "write binary" that indicates that the file is saved

import pickle
filename = "Python_program_Linear_Model.sav"
pickle.dump(Regressor, open(filename, "wb"))
```

```python
# The created model is then loaded using the "open" command. This command is in the class "load" in the "pickle" library.
# Note: "rb" means "read binary"
# The variable "Result" is then linked to the "predict" function in the class "load_model"
# A value of 13 is given to test the result of the model. It is noted that according to the model during the 13th year, the salary would be 148460
# LIkewise, a test carried out for the 15th year reavealed the salary as 167181
# The result is then called

load_model = pickle.load(open("Python_program_Linear_Model.sav","rb"))
Result = load_model.predict([[13]])
Result 
```
